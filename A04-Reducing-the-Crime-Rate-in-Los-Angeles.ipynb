### <b> 4. Data Cleaning and Data Transformation  </b>

##### <b> 4.1 Preprocessing  </b>


##### <b> 4.1.1 Feature Selection  </b>

We will remove irrelevant columns and amend column names. The selected features will be stored in ba775-a4.Crime_dataset.crime_set. 

```python
%%bigquery

CREATE OR REPLACE TABLE ba775-a4.Crime_dataset.crime_set AS
SELECT DR_NO AS Case_ID,  Date_Rptd AS Date_Reported,DATE_OCC AS Date_Occured,
TIME_OCC as Time_Occured, AREA_NAME as Area_Name, Crm_Cd as Crime_code,
Crm_Cd_Desc as Crime_Description,CAST(Vict_Age as STRING) as Victim_Age, 
Vict_Sex as Victim_Sex, Vict_Descent as Victim_Descent,
Status_Desc as Status_Description, Premis_Desc AS Premis_Description, 
Weapon_Desc_Status as Weapon_Description, Location, Cross_Street, LAT, LON
FROM `ba775-a4.Crime_dataset.Crime_Data_LA`;


SELECT *
FROM `ba775-a4.Crime_dataset.crime_set` LIMIT 5;
```

##### <b> 4.2 Data Quality Check  </b>

We used IF clause in each column to find null values. If the column has missing value, then the value '1' will be marked down. We sum result in each column to see the total number of missing values.


```python
%%bigquery

SELECT
   sum(IF(Case_ID is null,1,0)) as Case_ID_null,
   sum(IF(Date_Reported is null,1,0)) as Date_Reported_null,
   sum(IF(Date_Occured is null,1,0)) as Date_Occured_null,
   sum(IF(Time_Occured is null,1,0)) as Time_Occured_null,
   sum(IF(Area_Name is null,1,0)) as Area_Name_null,
   sum(IF(Crime_Description is null,1,0)) as Crime_Description_null,
   sum(IF(Victim_Age is null,1,0)) as Victim_Age_null,
   sum(IF(Victim_Sex is null,1,0)) as Victim_Sex_null,
   sum(IF(Victim_Descent is null,1,0)) as Victim_Descent_null,
   sum(IF(Weapon_Description is null,1,0)) as Weapon_Description_null,
   sum(IF(Status_Description is null,1,0)) as Status_Description_null,
   sum(IF(Location is null,1,0)) as Location_null,
   sum(IF(Cross_Street is null,1,0)) Cross_Street_null,
   sum(IF(LAT is null,1,0)) as LAT_null,
   sum(IF(LON is null,1,0)) as LON_null
FROM `ba775-a4.Crime_dataset.crime_set`
```

There are 75307 null values in Victim_Sex and 75312 null values in Victim_Descent. Also, there are 366774 and 472647 null values in Weapon_Description and Cross_Street, respectively. The null values in Victim_Sex and Victim_Descent may have occured because the victim's body was hard to identify, victim didn't want to disclose his/her identity  or, simply, because the reports were incomplete. The null values will be excluded by WHERE clause if needed.

##### <b> 4.3 Amending Records      </b>


We will amend records of certain columns to make the dataset easier to interpret and analyse.



##### <b> 4.3.1 Time_Occured      </b>

As time occurrences of crimes committed throughout the day is not insightful, we will categorise the 'Time_Occured' column into '12am to 6am', '6am to 12pm', '12pm to 6pm' and '6pm to 12am'. This will help us analyse and understand the different aspects of the data based on different parts of the day when crime was committed. 

```python
%%bigquery

UPDATE ba775-a4.Crime_dataset.crime_set 
SET Time_Occured = 
CASE
 WHEN CAST(Time_Occured AS int) BETWEEN 0001 AND 0559 THEN '12am to 6 am'
 WHEN CAST(Time_Occured AS int) BETWEEN 0600 AND 1159 THEN '6am to 12pm'
 WHEN CAST(Time_Occured AS int) BETWEEN 1200 AND 1759 THEN '12pm to 6pm'
 WHEN CAST(Time_Occured AS int) BETWEEN 1800 AND 2359 THEN '6pm to 12am'
ELSE 'Unknown' END 
where 1=1 
```

##### <b> 4.3.2 Victim_Descent      </b>

We will label the descents of the victims (as shown below) to make the data easier to understand. For example, for the sake of analysis, 'white' is much more informative than mere 'W'.

```python
%%bigquery
UPDATE ba775-a4.Crime_dataset.crime_set
SET Victim_Descent = 
CASE
  WHEN Victim_Descent = 'W' THEN 'white'
  WHEN Victim_Descent = 'A' THEN 'Other Asian'
  WHEN Victim_Descent = 'B' THEN 'Black'
  WHEN Victim_Descent = 'C' THEN 'Chinese'
  WHEN Victim_Descent = 'D' THEN 'Cambodian'
  WHEN Victim_Descent = 'F' THEN 'Filipino'
  WHEN Victim_Descent = 'G' THEN 'Guamanian'
  WHEN Victim_Descent = 'H' THEN 'Hispanic/Latin/Mexican'
  WHEN Victim_Descent = 'I' THEN 'American Indian/Alaskan Native'
  WHEN Victim_Descent = 'J' THEN 'Japanes'
  WHEN Victim_Descent = 'K' THEN 'Korean'
  WHEN Victim_Descent = 'L' THEN 'Laotian'
  WHEN Victim_Descent = 'O' THEN 'Other'
  WHEN Victim_Descent = 'P' THEN 'Pacific Islander'
  WHEN Victim_Descent = 'S' THEN 'Samoan'
  WHEN Victim_Descent = 'U' THEN 'Hawaiian'
  WHEN Victim_Descent = 'V' THEN 'Vietnamese'
  WHEN Victim_Descent = 'Z' THEN 'Asian Indian'
  WHEN Victim_Descent = 'X' THEN 'Unknown'
ELSE 'Unknown' END
where 1=1
```

##### <b> 4.3.3 Victim_Age      </b>

We will categorise the Victim Age into 'Infant', 'Children', 'Teenager', 'Young adults', 'Middle-aged',
'Old adults' and 'Seniors'. Doing so will help us understand the number and types of crimes committed on diffferent age groups.

```python
%%bigquery

UPDATE ba775-a4.Crime_dataset.crime_set
SET Victim_Age = 
    CASE 
    WHEN CAST(Victim_Age AS int) BETWEEN 0 AND 3 THEN 'Infant'
    WHEN CAST(Victim_Age AS int) BETWEEN  4 AND 12 THEN 'Children'
    WHEN CAST(Victim_Age AS int) BETWEEN 13 AND 19 THEN 'Teenager'
    WHEN CAST(Victim_Age AS int) BETWEEN 20 AND 29 THEN 'Young adults' 
    WHEN CAST(Victim_Age AS int) BETWEEN 30 AND 49 THEN 'Middle-aged' 
    WHEN CAST(Victim_Age AS int) BETWEEN  50 AND 64 THEN 'Old adults' 
    ELSE 'Seniors' END 
where 1=1
```

##### <b> 4.3.4 Victim_Sex      </b>

Victim_Sex column has irrelevant values such as X, H and NULL. As such, we will categorise the Victim Sex into 'Male', 'Female', and 'Unknown'. The 'Unknown' value represents those cases in which the victim's body was hard to identify, victim didn't want to disclose his/her identity or, simply, because the reports were incomplete.

```python
%%bigquery

UPDATE ba775-a4.Crime_dataset.crime_set
SET Victim_Sex = 
CASE
  WHEN Victim_Sex = 'M' THEN 'Male'
  WHEN Victim_Sex = 'F' THEN 'Female'
ELSE 'Unknown' END
where 1=1
```

##### <b> 4.3.5 Date_Occured and Date_Reported      </b>

The 'Date_Occured' and 'Date_Reported' columns contain '12:00:00 AM' after each record. As this information is repititive and not insightful, we will be removing it from the two columns. The 'Date_Occured' column contains the date on which the crime was committed, and 'Date_Reported' column contains the date on which the crime was reported. 


```python
%%bigquery
UPDATE  ba775-a4.Crime_dataset.crime_set
SET Date_Occured =  SUBSTRING(Date_Occured, 1, 10)
where 1=1;

UPDATE  ba775-a4.Crime_dataset.crime_set
SET Date_Reported =  SUBSTRING(Date_Reported, 1, 10)
where 1=1;
```

##### <b> 4.3.6 Location and Cross_Street      </b>

'Cross_Street' column contains additional details of the location where the crime was committed. 

As mentioned in the data quality check, Cross_Street column has a large number of null values (472647 null values). As Cross_Street is a drilled-down detail of the Location of the crime committed, we will include the non-null values of Cross_Street in the Location column. The Cross_Street column is no longer needed, we will remove it.

```python
%%bigquery
UPDATE ba775-a4.Crime_dataset.crime_set
SET Location = CONCAT(ifnull(Location,''),' ',ifnull(Cross_Street,''))
where 1=1;

ALTER TABLE  ba775-a4.Crime_dataset.crime_set
DROP COLUMN Cross_Street; 
```

**[Question 1.4] From 2020 to 2022, which year has the highest number of cases?**

```python
%%bigquery
SELECT Year, Number_of_Cases, ROUND((Number_of_Cases/SUM(Number_of_Cases) OVER()*100), 2) AS Number_of_Cases_Percentage
FROM
(
SELECT Year, COUNT(Case_ID) AS Number_of_Cases
FROM
  (
    Select Case_ID, Extract(year From (PARSE_DATE('%m/%d/%Y', Date_Reported))) AS Year
    From ba775-a4.Crime_dataset.crime_set
    where Extract(month From (PARSE_DATE('%m/%d/%Y', Date_Reported))) BETWEEN 1 and 8
  )
GROUP BY Year
)
ORDER BY Number_of_Cases DESC;
```

2022 has the highest number of cases whereas 2020 has lowest and the crime cases are increasing as we move from 2020 to 2022.
As the data is available from 1 Jan 2020 to 12 Sep 2022, it won't be a fair comparison to include all the months of 2020 and 2021. Therefore, we have compared the crime cases in the first 8 months (Jan - Aug) of each year.


**[Question 1.5a] From 2020 to 2022, analyse the month-to-month crime cases from January to August?**

```python
%%bigquery
SELECT Month, year, Number_of_Cases, ROUND((Number_of_Cases/SUM(Number_of_Cases) OVER(PARTITION BY Month)*100), 2) AS Number_of_Cases_Percentage
FROM
(
 SELECT EXTRACT(month FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) as Month,EXTRACT(year FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) as Year, COUNT(Case_ID) as Number_of_Cases,
 FROM `ba775-a4.Crime_dataset.crime_set`
 GROUP BY month, year
)
WHERE Month IN (1, 2, 3, 4, 5, 6, 7, 8)
ORDER BY month, year DESC
```

From January to August, the monthly crime cases are generally increasing as we move from 2020 to 2022.

**[Question 1.5b] From 2020 to 2021, analyse the month-to-month cases from September to December?**

```python
%%bigquery
SELECT Month, year, Number_of_Cases, ROUND((Number_of_Cases/SUM(Number_of_Cases) OVER(PARTITION BY Month)*100), 2) AS Number_of_Cases_Percentage
FROM
(
 SELECT EXTRACT(month FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) as Month, EXTRACT(year FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) as Year, COUNT(Case_ID) as Number_of_Cases
 FROM `ba775-a4.Crime_dataset.crime_set`
 GROUP BY month, year
)
WHERE Year IN (2020,2021) AND Month IN (9,10,11,12)
ORDER BY month, year DESC
LIMIT 10;
```

Comparing the 2020 and 2021 cases from September to December, we can clearly see that the cases in 2021 are higher than those in 2020.

**[Question 4.3] Which months have the highest crime rate in LA?**

```python
%%bigquery
SELECT Month, Cases, ROUND((Cases/SUM(Cases) OVER()*100), 2) AS Cases_Percentage
FROM
(
SELECT EXTRACT(month FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) as Month, count(Case_ID) as Cases
FROM `ba775-a4.Crime_dataset.crime_set`
WHERE EXTRACT(year FROM PARSE_DATE('%m/%d/%Y', Date_Occured)) IN (2020,2021)
GROUP BY Month
)
ORDER BY Cases DESC 
LIMIT 5;
```

Since we have data till 12 September 2022, we have narrowed down our analysis to 2020 and 2021.
According to the analysis, the months of July and October have the highest crime rate, followed by August. 


**[Question 4.4]  For cases with 'Invest Cont' status (cases that haven't reached a conclusion yet), find the number of such cases and the days that have already passed by for these cases?** 


```python
%%bigquery
SELECT Time_Passed_By, SUM(Number_of_Cases_Pending) as Number_of_Pending_Cases
FROM
(  Select CASE
    WHEN Days_Already_Passed_By BETWEEN 1 AND 180 THEN 'Upto 6 months'
    WHEN Days_Already_Passed_By BETWEEN 181 AND 365 THEN '6 Months to 1 Year'
    WHEN Days_Already_Passed_By BETWEEN 366 AND 731 THEN '1 to 2 Years'
    WHEN Days_Already_Passed_By >= 732 THEN 'More than 2 Years' END AS Time_Passed_By, 
  COUNT(Case_ID) AS Number_of_Cases_Pending
  FROM
    (
      SELECT Case_ID, Status_Description, DATE_DIFF(CURRENT_DATE(),
      CAST(PARSE_DATE('%m/%d/%Y', Date_Reported) AS date),day) as Days_Already_Passed_By
      FROM ba775-a4.Crime_dataset.crime_set
      WHERE Status_Description = "Invest Cont"
    )
  group by Days_Already_Passed_By
)
group by Time_Passed_By
order by Time_Passed_By;
```

The above result shows the number of cases pending and the time already passed by. Surprisingly, there are large number of cases which haven't been solved or reached a conclusion yet. 

As such, for cases for which 2 years have already passed since reported, there are as many as 116159 unsolved cases (20.42% of the total cases).
